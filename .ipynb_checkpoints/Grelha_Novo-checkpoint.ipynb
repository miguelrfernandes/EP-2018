```python
class Individuo:
    
    def __init__(self,ID,estado):
        self._ID=ID
        self._estado=estado
        
    def ID(self):
        return self._ID
    
    def estado(self):
        return self._estado
    
    def mostraI(self):
        print ('ID:', self._ID,'  ','Estado:', self._estado)
```

```python
import random

class Grelha:
    
################################################# PARA APAGAR ###########################################
    
    def mostraG(self):
        print (len(self.G)*' _')
        for i in self.G:
            for j in i:
                print('|', end="")
                if j==None:
                    print ('_',end="")
                elif j=='O':
                    print ('\u25AA',end="")
                else:
                    print (j._estado,end="")
            print ('|')
            
    def verestados(self):
        for i in range(len(self.G)+2):
            print("#", end=" ")
        print("")
        for i in self.G:
            print("#", end=" ")
            for j in i:
                if j=="O":
                    print("*", end=" ")
                elif j==None:
                    print("-", end=" ")
                else:
                    print(j.estado(), end=" ")
            print("#")
        for i in range(len(self.G)+2):
            print("#", end=" ")
            
    def mostraIDG(self):
        print (len(self.G)*' _____')
        for i in self.G:
            for j in i:
                print('|', end='')
                if j=='O':
                    print ('','\u25AA','\u25AA',end=' ')
                elif j==None:
                    print ('_____',end='')
                else:  
                    if j._ID<=9:
                        print (' ',j._ID,end='  ')
                    if 9<j._ID<100:
                        print (' ',j._ID,end=' ')
                    if j._ID>=100:
                        print ('',j._ID,end=' ')
            print ('|')
    
################################################# PARA APAGAR ###########################################
    
    def __init__(self,N,Ps,Pi,o,cen):
        ID=iter(range(1,4*(N**2)+4*N+2))
        self._N=N
        self._l=2*N+1
        
        w=[]
        for j in range(2*N+1):
            s=[]
            for i in range(2*N+1):
                s=s+[None]
            w=w+[s]
        
        for p in o:
            w[N-p[1]][N+p[0]]='O'
        
        if cen==1:
            while Ps>0:
                a=random.randint(-N,N)
                b=random.randint(-N,N)
                if w[N-b][N+a]==None:
                    w[N-b][N+a]=Individuo(next(ID),'S')
                    Ps=Ps-1
                
            while Pi>0:
                a=random.randint(-N,N)
                b=random.randint(-N,N)
                if w[N-b][N+a]==None:
                    w[N-b][N+a]=Individuo(next(ID),'I')
                    Pi=Pi-1
                
        #if cen==b:
            #Fazer
             
        self.G=w 
        
    def insI(self,ind,p):
        self.G[self._N-p[1]][self._N+p[0]]=ind
    
    def delI(self,p):
        self.G[self._N-p[1]][self._N+p[0]]=None
        
    def livreQ(self,p):
        return self.G[self._N-p[1]][self._N+p[0]]==None
    
    def encontrar(self,ID):
        for i in range(len(self.G)):
            for j in range(len(self.G)):
                if self.G[i][j]!="O" and self.G[i][j]!=None:
                    if self.G[i][j]._ID==ID:
                        return [j-self._N,self._N-i]
    
    def nInf(self):
        n=0
        for i in self.G:
            for j in i:
                if j is not None and not isinstance(j,str):
                    if j._estado=='I':
                        n=n+1
        return n   
    
    def c_estados(self):
        w=[[],[],[],[]]
        i=0
        while i<len(self.G):
            j=0
            while j<len(self.G):
                if self.G[i][j] is not None and not isinstance(self.G[i][j],str):
                    if self.G[i][j]._estado=='S':
                        w[0]=w[0]+[(j-self._N,self._N-i)]
                    if self.G[i][j]._estado=='E':
                        w[1]=w[1]+[(j-self._N,self._N-i)]
                    if self.G[i][j]._estado=='I':
                        w[2]=w[2]+[(j-self._N,self._N-i)]
                    if self.G[i][j]._estado=='R':
                        w[3]=w[3]+[(j-self._N,self._N-i)]
                j+=1
            i+=1
        return w
```
